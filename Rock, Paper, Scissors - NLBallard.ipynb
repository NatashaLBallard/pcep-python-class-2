Create a program where the user can play a game of rock, paper, scissors against a computer. Set a play choice (rock, paper, or scissors) for the computer player and during the program, request for the user to input their play choice.

The rules in Rock, Paper, Scissors are:

- Rock beats scissors
- Scissors beats paper
- Paper beats rock

If the user wins, then display "You win!" If they lose against the computer, then display "You lose." If the computer and the user have the same play choice, display "It's a draw."

BONUS: Randomize the play choice for the computer (hint: Python has a library called random). Also ask the user if they would like to play again. If they do, start the game again but if not, thank the user for playing the game.

```python
rock = "r"
paper = "p"
scissors = "s"
#choices = ('r','p','s')
numOfChoice = {
  
    1:{'r':'rock'},
    2:{'p':'paper'},
    3:{'s':'scissors'},
}
scores = {'Wins':0, 'Losses':0}


import random as rand
compChoice = 'r'
#compChoice = rand.randint(0,3)





userChoice = input("Your Turn!  Rock, Paper, or Scissors? Choose: (r/p/s) ")

def play(userChoice) :
    if userChoice == 'r':
        if compChoice == 's': 
            print("User Wins")
            scores['Wins'] += 1
        elif compChoice == 'p':
            print("User Loses")
            scores['Losses'] += 1
        else :
            print("Result: Draw") 
            
    if userChoice == 's':
        if compChoice == 'p':
            print("User Wins")
            scores['Wins'] += 1
        elif compChoice == 'r':
            print("User Loses")
            scores['Losses'] += 1
        else :
            print("Result: Draw") 

    if userChoice == 'p':
        if compChoice == 'r':
            print("User Wins")
            scores['Wins'] += 1
        elif compChoice == 's':
            print("User Loses")
            scores['Losses'] += 1
        else :
            print("Result: Draw") 
                    





#print(f'Computer Choice: {numOfChoice[compChoice].values()} \t User Choice: {userChoice}')

play(userChoice)
print(f'Computer Choice: {compChoice} \t User Choice: {userChoice}')
print(scores)

```
